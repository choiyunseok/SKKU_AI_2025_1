## 1) nn.Conv2d
- Creates a convolution layer for 2-dimensional data.
- When you set the necessary hyperparameters, PyTorch automatically defines the parameters (weights) for the convolution layer.
- `Conv2d(in_channels, out_channels, kernel_size, stride=1, padding=0, dilation=1, groups=1, bias=True, padding_mode='zeros')`

### parameters
  - in_channels: Number of input channels. For grayscale images, use 1; for RGB images, use 3.
  - out_channels: Number of output channels.
  - kernel_size: Size of the convolution kernel.
  - stride: Stride size.
  - padding: Padding size.

### shape
1. Input tensor($N, C_{in}, H_{in}, W_{in}$)
 - $N$: : Batch size
 - $C_{in}$ : Must match the value specified in `in_channerls`
 - $H_{in}$ :  Height of the 2D input tensor
 - $ W_{in}$:  Width of the 2D input tensor


- [doc] (https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html)

```python
import torch
import torch.nn as nn
```

### Example 1) kernel_size=3, stride=1, padding=0
![image.png](http://drive.google.com/uc?id=1dvnKVDptCOD1VLtiiTztOB2Lp4SAkuOE)

```python
input = torch.ones(1,1,4,4) # Batch_size * Channel * Height * Width
conv_layer = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=3, stride=1, padding=0)
output = conv_layer(input)
print(output.shape)
```
